Duration percentage and plot for each categories 

```python
import os
from moviepy import VideoFileClip
import matplotlib.pyplot as plt
import paths
from ansi_colors import *



directory = paths.DATASET
duration_salsa = 0
duration_bachata = 0


for filename in os.listdir(directory):
    if filename.endswith(".mp4"):
        video_path = os.path.join(directory, filename)

        video = VideoFileClip(video_path)

        duration = video.duration

        if filename.startswith("salsa_"):
            duration_salsa += duration
        elif filename.startswith("bachata_"):
            duration_bachata += duration

categories = ["salsa", "bachata"]
durations = [duration_salsa, duration_bachata]

total_duration = duration_bachata + duration_salsa

print(f"\n{GREEN}PERCENTAGE:{(duration_bachata/ total_duration) * 100}% Bachata, {(duration_salsa/ total_duration) * 100}% Salsa\n{RESET}")



plt.bar(categories, durations, color=['blue', 'orange'])
plt.title("Video duration")
plt.xlabel("Class")
plt.ylabel("Duration (s)")
plt.show()
```
